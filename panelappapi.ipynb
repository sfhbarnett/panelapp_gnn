```python
import tqdm
import time
import json

gene2panel = {}

for gene in tqdm.tqdm(list(set(genes)), total=len(set(genes))):
    if gene in gene2panel.keys():
        continue
    time.sleep(0.1)  # Avoid hitting the API too hard
    r = requests.get(f"https://panelapp.genomicsengland.co.uk/api/v1/entities/{gene}/")
    panels = []
    for x in r.json()['results']:
        panels.append(x['panel']['id'])
    gene2panel[gene] = panels
    with open(f"gene_data/{gene}.json", "w") as f:
        json.dump(r.json(), f)
```

```python
import torch
from torch_geometric.nn import HeteroConv, SAGEConv

class GenePanelGNN(torch.nn.Module):
    def __init__(self, hidden=128):
        super().__init__()
        self.conv1 = HeteroConv({
            ('gene','in','panel'):  SAGEConv((-1, -1), hidden),
            ('panel','rev_in','gene'): SAGEConv((-1, -1), hidden)
        }, aggr='mean')
        self.conv2 = HeteroConv({
            ('gene','in','panel'):  SAGEConv((-1, -1), hidden),
            ('panel','rev_in','gene'): SAGEConv((-1, -1), hidden)
        }, aggr='mean')
        self.score = torch.nn.Bilinear(hidden, hidden, 1)   # link predictor

    def forward(self, x_dict, edge_index_dict, pairs):
        x_dict = self.conv1(x_dict, edge_index_dict)
        x_dict = {k: x.relu() for k,x in x_dict.items()}
        x_dict = self.conv2(x_dict, edge_index_dict)
        g_emb = x_dict['gene'][pairs[0]]
        p_emb = x_dict['panel'][pairs[1]]
        return self.score(g_emb, p_emb).squeeze(-1)
```

```python
import torch.nn.functional as F
from torch_geometric.utils import negative_sampling
pos_edge_index = data['gene','in','panel'].edge_index  # known links
neg_sampler   = torch_geometric.utils.negative_sampling

E = 100
model = GenePanelGNN(hidden=128)
n_genes = data['gene'].num_nodes
n_panels = data['panel'].num_nodes
x = data.x_dict  # node features
optimiser = torch.optim.Adam(model.parameters(), lr=1e-3)


x_dict = {            # node-feature dictionaries
    'gene' : data['gene'].x,
    'panel': data['panel'].x
}
edge_index_dict = {   # connectivity dictionaries
    ('gene','in','panel')    : data['gene','in','panel'].edge_index,
    ('panel','rev_in','gene'): data['panel','rev_in','gene'].edge_index
}

pos_score = model(x_dict, edge_index_dict, pos_edge_index)  # works
neg_score = model(x_dict, edge_index_dict, neg_edges)

pos_edge_index = data['gene','in','panel'].edge_index        # (2, E)
n_genes, n_panels = data['gene'].num_nodes, data['panel'].num_nodes

for epoch in range(E):
    
    neg_edges = neg_sampler(
        pos_edge_index, num_nodes=(n_genes, n_panels),
        num_neg_samples=pos_edge_index.size(1)
    )

    pos_score = model(x_dict, edge_index_dict, pos_edge_index)
    neg_score = model(x_dict, edge_index_dict, neg_edges)

    loss = F.binary_cross_entropy_with_logits(
        torch.cat([pos_score, neg_score]),
        torch.cat([torch.ones_like(pos_score),
                   torch.zeros_like(neg_score)])
    )
    loss.backward(); optimiser.step(); optimiser.zero_grad()
    print(f"Epoch {epoch+1}/{E}, Loss: {loss.item():.4f}")
```

```python
def ensure_gene(gene_symbol, feats):
    if gene_symbol not in gene2idx:           # brand-new gene
        gid = len(gene2idx)
        gene2idx[gene_symbol] = gid
        # append its feature vector (or a learnable embed if feats is None)
        f = feats if feats is not None else torch.zeros(1, data['gene'].x.size(1))
        data['gene'].x = torch.cat([data['gene'].x, f], dim=0)
    return gene2idx[gene_symbol]

def add_edge(gid, pid):
    ei = data['gene','in','panel'].edge_index
    new = torch.tensor([[gid],[pid]])
    data['gene','in','panel'].edge_index = torch.cat([ei, new], dim=1)
    data['panel','rev_in','gene'].edge_index = torch.cat([ei.flip(0), new.flip(0)], dim=1)

def suggest_panels_for_gene(gene_symbol, known_panels, k=10):
    gid = gene2idx[gene_symbol]
    all_pids = torch.arange(data['panel'].num_nodes)

    # drop panels the curators have already added
    known_pids = torch.tensor([panel2idx[p] for p in known_panels])
    mask = ~torch.isin(all_pids, known_pids)
    cand_pids = all_pids[mask]

    # build (gene, panel) pair tensor
    pairs = (torch.full_like(cand_pids, gid), cand_pids)

    # score with frozen model
    scores = torch.sigmoid(model(x_dict, edge_index_dict, pairs))
    top = torch.topk(scores, k=min(k, len(cand_pids))).indices
    return [int(idx2panel[cand_pids[i].item()]) for i in top]
```

```python
# ------------------------------------------
# 1. utilities
# ------------------------------------------
def ensure_gene(gene_symbol: str,
                feat: torch.Tensor | None = None) -> int:
    """
    Make sure `gene_symbol` exists in the graph.
    Returns its internal index (gid).
    """
    gid = gene2idx.get(gene_symbol)
    if gid is None:                                  # brand-new
        gid = len(gene2idx)
        gene2idx[gene_symbol] = gid

        # append a feature row
        if feat is None:           # cheap fallback: small random vector
            feat = torch.randn(data['gene'].x.size(1)) * 0.01
        data['gene'].x = torch.cat([data['gene'].x, feat[None]], 0)
    return gid


def add_edge(gid: int, pid: int):
    """
    Append one (gene gid → panel pid) edge (+ reverse) to the graph.
    """
    e = torch.tensor([[gid], [pid]])
    data['gene', 'in', 'panel'   ].edge_index = torch.cat(
        [data['gene', 'in', 'panel'   ].edge_index, e], dim=1)
    data['panel', 'rev_in', 'gene'].edge_index = torch.cat(
        [data['panel', 'rev_in', 'gene'].edge_index, e.flip(0)], dim=1)


# ------------------------------------------
# 2. add a new gene–panel link
# ------------------------------------------
gene   = 'MYNEWGENE'       # HGNC symbol just entered by a curator
panel  = 285               # numeric PanelApp id (e.g. “Hereditary breast cancer”)

gid = ensure_gene(gene)              # create/lookup the gene
pid = panel2idx[panel]               # panels were pre-indexed
add_edge(gid, pid)                   # store the new membership

# ------------------------------------------
# 3. build fresh inputs for the model
# ------------------------------------------
x_dict = {
    'gene' : data['gene'].x,
    'panel': data['panel'].x,
}
edge_index_dict = {
    ('gene',  'in',     'panel'): data['gene',  'in',     'panel'].edge_index,
    ('panel', 'rev_in', 'gene' ): data['panel', 'rev_in', 'gene' ].edge_index,
}

# ------------------------------------------
# 4. score every *other* panel for this gene
# ------------------------------------------
all_pids  = torch.arange(data['panel'].num_nodes)
mask      = all_pids != pid              # don’t propose the one we just added
cand_pids = all_pids[mask]
pairs     = (torch.full_like(cand_pids, gid), cand_pids)

with torch.no_grad():
    scores = torch.sigmoid(model(x_dict, edge_index_dict, pairs))

top = torch.topk(scores, k=10).indices          # best 10
suggestions = pd.DataFrame({
    "panel_id": [int(idx2panel[cand_pids[i].item()]) for i in top],
    "score": [round(float(scores[i]), 3) for i in top]
})
# suggestions = "".join([f"{str(int(idx2panel[int(cand_pids[i])]))}, {str(round(float(scores[i]),3))}\n" for i in top])

print("Suggested extra panels for", gene, ":\n", suggestions)
```
